```python

def plot_correlation_heatmap(data, region):
    plt.figure(figsize=(8, 6))
    
    
    correlation_matrix = data.corr(numeric_only=True)
    
    
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
    
    plt.title(f"Correlation Heatmap - {region}")
    plt.show()


    
    
#plot_correlation_heatmap(df_na, "North America")   # These are not meaningful heat maps because 
#plot_correlation_heatmap(df_la, "Latin America")             # only considering two countries.
plot_correlation_heatmap(df_eu, "Europe")
plot_correlation_heatmap(df_ap, "Asia-Pacific")
```

```python


def run_regression(data, region):
    
    # Define independent are cultural measures and dependent are Coca-Cola consumption variables.
    X = data[['PDI', 'IDV', 'MTS', 'UAI', 'LTO', 'IVR']]
    y = data['Consumption of Coke in Millions of Liters 2024']
    

    X = sm.add_constant(X)
    

    model = sm.OLS(y, X).fit()
    

    print(f"\nRegression Results for {region}:\n")
    print(model.summary())

# Run regression for each region
run_regression(df_na, "North America")
run_regression(df_la, "Latin America")
run_regression(df_eu, "Europe")
run_regression(df_ap, "Asia-Pacific")
```

```python

def plot_scatter(data, region):
    cultural_vars = ['PDI', 'IDV', 'MTS', 'UAI', 'LTO', 'IVR']
    
    for var in cultural_vars:
        plt.figure(figsize=(6, 4))
        sns.scatterplot(x=data[var], y=data['Consumption of Coke in Millions of Liters 2024'])
        plt.xlabel(var)
        plt.ylabel('Coca-Cola Consumption')
        plt.title(f'{var} vs. Coca-Cola Consumption ({region})')
        plt.show()

# each region
plot_scatter(df_na, "North America")
plot_scatter(df_la, "Latin America")
plot_scatter(df_eu, "Europe")
plot_scatter(df_ap, "Asia-Pacific")
```

```python
features_culture = ['PDI', 'IDV', 'MTS', 'UAI', 'LTO', 'IVR']



for sheet_name, df in sheets.items():
    if set(features_culture).issubset(df.columns):
        num_samples = df.shape[0]
        num_clusters = min(3, num_samples)  

        if num_clusters > 1:
            
            scaler = StandardScaler()
            df_scaled = scaler.fit_transform(df[features_culture])

            
            kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
            df['Cluster'] = kmeans.fit_predict(df_scaled)

           
            plt.figure(figsize=(15, 10))
            for i, dimension in enumerate(features_culture, 1):
                plt.subplot(2, 3, i)
                sns.scatterplot(data=df, x='Consumption of Coke in Millions of Liters 2024', 
                                y=dimension, hue='Cluster', palette='viridis', style='Cluster', s=100)
                plt.title(f'{sheet_name}: {dimension} vs Coca-Cola Consumption')
                plt.xlabel('Coca-Cola Consumption (Millions of Liters, 2024)')
                plt.ylabel(dimension)

            plt.suptitle(f'K-Means Clustering in {sheet_name}', fontsize=16)
            plt.tight_layout(rect=[0, 0, 1, 0.96])
            plt.show()


            print(f"Updated Data for {sheet_name}:")
            print(df[['Country', 'Consumption of Coke in Millions of Liters 2024', 'Cluster']].head())
        else:
            print(f"Not enough data points for clustering in {sheet_name}, skipping.")

    else:
        print(f"Missing required columns in {sheet_name}, skipping.")
```

```python
features_income = ['AVG Monthly Income', 'AVG Annual Income']


for sheet_name, df in sheets.items():
    if set(features_income).issubset(df.columns):
        num_samples = df.shape[0]
        num_clusters = min(3, num_samples)  

        if num_clusters > 1:

            scaler = StandardScaler()
            df_scaled = scaler.fit_transform(df[features_income])


            kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
            df['Cluster'] = kmeans.fit_predict(df_scaled)


            plt.figure(figsize=(15, 10))
            for i, dimension in enumerate(features_income, 1):
                plt.subplot(1, 2, i)
                sns.scatterplot(data=df, x='Consumption of Coke in Millions of Liters 2024', 
                                y=dimension, hue='Cluster', palette='viridis', style='Cluster', s=100)
                plt.title(f'{sheet_name}: {dimension} vs Coca-Cola Consumption')
                plt.xlabel('Coca-Cola Consumption (Millions of Liters, 2024)')
                plt.ylabel(dimension)

            plt.suptitle(f'K-Means Clustering by Income in {sheet_name}', fontsize=16)
            plt.tight_layout(rect=[0, 0, 1, 0.96])
            plt.show()


            print(f"Updated Data for {sheet_name}:")
            print(df[['Country', 'Consumption of Coke in Millions of Liters 2024', 'Cluster']].head())
        else:
            print(f"Not enough data points for clustering in {sheet_name}, skipping.")

    else:
        print(f"Missing required columns in {sheet_name}, skipping.")
```

```python
###

## Code for all the bar graphs

### 


file_path_2 = "Coke_Data_Data400.xlsx" 
excel_data = pd.ExcelFile(file_path)


all_data = []

# By sheet
for sheet_name in excel_data.sheet_names:
    df = excel_data.parse(sheet_name)
    
    
    if 'Country' in df.columns and 'IDV' in df.columns:
        sheet_data = df[['Country', 'IDV']]
        sheet_data['Sheet'] = sheet_name  
        all_data.append(sheet_data)


combined_data = pd.concat(all_data, ignore_index=True)


plt.figure(figsize=(12, 6))
plt.bar(combined_data['Country'] + " (" + combined_data['Sheet'] + ")", combined_data['IDV'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('Country (Sheet Name)')
plt.ylabel('IDV Score')
plt.title('IDV Scores for Countries Across Sheets')
plt.tight_layout()
plt.show()
```
