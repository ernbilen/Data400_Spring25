```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

#file_path = r"C:\Users\grace\OneDrive\Documents\Data400\Baseball League Stats1.xlsx"
#dfs = pd.read_excel(file_path, sheet_name=None, engine="openpyxl")
#print("Sheet names:", dfs.keys())

# Get batting, pitching, and monthly data
batting = dfs["Batting"]
pitching = dfs["Pitching"]
monthly = dfs["WinsLoss by Month"]

# Drop playoff-related columns from batting and pitching
batting = batting.drop(columns=[col for col in batting.columns if 'Playoffs' in col], errors='ignore')
pitching = pitching.drop(columns=[col for col in pitching.columns if 'Playoffs' in col], errors='ignore')

# Fix the column name difference: rename 'Teams' to 'Team' in the pitching dataframe
if 'Teams' in pitching.columns and 'Team' not in pitching.columns:
    pitching = pitching.rename(columns={'Teams': 'Team'})
    print("Renamed 'Teams' column to 'Team' in pitching data")

# Print column names to verify
print("Batting columns:", batting.columns.tolist()[:5], "...")
print("Pitching columns:", pitching.columns.tolist()[:5], "...")
print("Monthly columns:", monthly.columns.tolist())

# Remove the "MLB" average row 
batting = batting[batting['Team'] != 'MLB']
pitching = pitching[pitching['Team'] != 'MLB']
monthly = monthly[monthly['Team'] != 'MLB'] if 'Team' in monthly.columns else monthly

# Merge batting and pitching data 
merged_data = pd.merge(batting, pitching, on=['Team', 'Season'], suffixes=('_batting', '_pitching'))

# Handle duplicate 'Playoffs' column after merge
if 'Playoffs_batting' in merged_data.columns and 'Playoffs_pitching' in merged_data.columns:
    # Check if they're the same
    if (merged_data['Playoffs_batting'] == merged_data['Playoffs_pitching']).all():
        merged_data['Playoffs'] = merged_data['Playoffs_batting']
    else:
        # If there are discrepancies, prioritize batting data
        print("Warning: Discrepancies found between batting and pitching playoff data. Using batting data.")
        merged_data['Playoffs'] = merged_data['Playoffs_batting']
    
    # Drop the duplicates
    merged_data = merged_data.drop(['Playoffs_batting', 'Playoffs_pitching'], axis=1)
elif 'Playoffs_batting' in merged_data.columns:
    merged_data = merged_data.rename(columns={'Playoffs_batting': 'Playoffs'})
elif 'Playoffs_pitching' in merged_data.columns:
    merged_data = merged_data.rename(columns={'Playoffs_pitching': 'Playoffs'})
elif 'Playoffs?' in merged_data.columns:
    merged_data = merged_data.rename(columns={'Playoffs?': 'Playoffs'})

# Now handle the 'Playoffs?' column if it exists instead of 'Playoffs'
if 'Playoffs?' in merged_data.columns and 'Playoffs' not in merged_data.columns:
    merged_data = merged_data.rename(columns={'Playoffs?': 'Playoffs'})

# Check if we need to merge with monthly data
if 'Team' in monthly.columns and 'Season' in monthly.columns:
    # Group the monthly data by Team and Season (aggregate wins and losses)
    monthly_aggregated = monthly.groupby(['Team', 'Season']).agg({
        'Wins': 'sum',
        'Loss': 'sum',
        'Playoffs': 'first'  
    }).reset_index()
    
    # Rename columns to avoid conflicts
    monthly_aggregated = monthly_aggregated.rename(columns={
        'Wins': 'Total_Wins',
        'Loss': 'Total_Losses'
    })
    
    # Merge with monthly aggregated data
    merged_data = pd.merge(merged_data, monthly_aggregated, on=['Team', 'Season'], suffixes=('', '_monthly'))
    print("Monthly data aggregated and merged successfully")
else:
    print("Note: WinsLoss by Month sheet doesn't have expected columns for merging.")

# Verify the merge worked correctly
print(f"Merged data shape: {merged_data.shape}")
print(f"Playoffs column in merged data: {'Playoffs' in merged_data.columns}")

# Handle the case where 'Playoffs' might be 'Playoffs?'
if 'Playoffs?' in merged_data.columns and 'Playoffs' not in merged_data.columns:
    merged_data = merged_data.rename(columns={'Playoffs?': 'Playoffs'})

# Make sure Playoffs is available
if 'Playoffs' not in merged_data.columns:
    print("Error: No 'Playoffs' column found in merged data. Please check your input data.")
    # Create a dummy column as a fallback
    merged_data['Playoffs'] = 0
else:
    print(f"Teams that made playoffs: {merged_data['Playoffs'].sum()}")
    print(f"Teams that didn't make playoffs: {len(merged_data) - merged_data['Playoffs'].sum()}")

# Convert 'Playoffs' to numeric if it's not already
merged_data['Playoffs'] = pd.to_numeric(merged_data['Playoffs'], errors='coerce').fillna(0).astype(int)

# Prepare features and target
cols_to_drop = ['Team', 'Season', 'Playoffs']
# Remove any other columns that shouldn't be used as features
X = merged_data.drop(cols_to_drop, axis=1)
y = merged_data['Playoffs']

# Print feature columns to verify
print(f"Number of features used: {X.shape[1]}")
print("First few feature columns:", X.columns.tolist()[:5], "...")

# Handle NaN values
X = X.fillna(X.mean())

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

# Scale numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train models
# 1. Logistic Regression
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train_scaled, y_train)

# 2. Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train) 

# 3. Gradient Boosting
gb = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb.fit(X_train, y_train) 

# Evaluate models
models = {
    'Logistic Regression': (log_reg, X_test_scaled),
    'Random Forest': (rf, X_test),
    'Gradient Boosting': (gb, X_test)
}

# Function to evaluate and print results
def evaluate_model(name, model, X_test_data, y_test_data):
    # Make predictions
    y_pred = model.predict(X_test_data)
    y_prob = model.predict_proba(X_test_data)[:, 1]
    
    # Print classification report
    print(f"\nModel: {name}")
    print(classification_report(y_test_data, y_pred))
    print(f"ROC-AUC Score: {roc_auc_score(y_test_data, y_prob):.4f}")
    
    # Print confusion matrix
    cm = confusion_matrix(y_test_data, y_pred)
    print("Confusion Matrix:")
    print(cm)
    
    # Return feature importances if available
    if hasattr(model, 'feature_importances_'):
        return model.feature_importances_
    elif hasattr(model, 'coef_'):
        return model.coef_[0]
    else:
        return None

# Evaluate all models
feature_importances = {}
for name, (model, X_test_data) in models.items():
    importances = evaluate_model(name, model, X_test_data, y_test)
    if importances is not None:
        feature_importances[name] = importances

# Plot feature importances for Random Forest
if 'Random Forest' in feature_importances:
    plt.figure(figsize=(12, 8))
    features = X.columns
    importances = feature_importances['Random Forest']
    indices = np.argsort(importances)[::-1]
    
    plt.title('Feature Importances (Random Forest)')
    plt.barh(range(min(20, len(indices))), importances[indices][:20])  
    plt.yticks(range(min(20, len(indices))), [features[i] for i in indices][:20])
    plt.tight_layout()
    plt.show()

# Function to find thresholds
def find_thresholds(model, X_data, feature_names):
    # For each feature, find the threshold that best separates playoff from non-playoff teams
    thresholds = {}
    
    # Get feature importances from our trained random forest
    if hasattr(model, 'feature_importances_'):
        importances = model.feature_importances_
    else:
        return thresholds
    
    # Sort features by importance
    sorted_idx = np.argsort(importances)[::-1]
    top_features = [feature_names[i] for i in sorted_idx[:10]] 
    
    # For each important feature
    for feature in top_features:
        # Create a simple decision tree to find the threshold
        dt = DecisionTreeClassifier(max_depth=1)
        
        # Extract just this feature as a DataFrame column
        X_feature = X_test[[feature]]
        dt.fit(X_feature, y_test)
        
        # Get the threshold
        if len(dt.tree_.threshold) > 0 and dt.tree_.threshold[0] != -2:
            threshold = dt.tree_.threshold[0]
            thresholds[feature] = threshold
    
    return thresholds

# Find thresholds for important features
thresholds = find_thresholds(rf, X_test, X.columns)
print("\nKey statistic thresholds for playoff prediction:")
for feature, threshold in sorted(thresholds.items(), key=lambda x: feature_importances['Random Forest'][list(X.columns).index(x[0])], reverse=True):
    print(f"{feature}: {threshold:.4f}")

# Calculate playoff probability for each team
print("\nPlayoff Probabilities by Team:")
for index, row in merged_data.iterrows():
    team_name = row['Team']
    season = row['Season']
    actual_playoff = row['Playoffs']
    
    # Get just the feature columns 
    team_features = row[X.columns].values.reshape(1, -1)
    
    # Calculate probability using the Random Forest model
    playoff_proba = rf.predict_proba(team_features)[0][1]
    
    print(f"{team_name} ({season}): {playoff_proba:.2%} chance of playoffs (Actual: {'Yes' if actual_playoff == 1 else 'No'})")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
local_path = r"C:\Users\grace\OneDrive\Documents\Data400\Baseball League Stats1.xlsx"
dfs = pd.read_excel(local_path, sheet_name=None, engine="openpyxl")
monthly_data = dfs["WinsLoss by Month"]

# Clean data if necessary
monthly_data = monthly_data[monthly_data['Team'] != 'MLB']  # Remove MLB averages if present

# Create additional features that might be predictive
monthly_data['Win_Pct'] = monthly_data['Wins'] / (monthly_data['Wins'] + monthly_data['Loss'])

# Group by Team and Season, then create features for each month
pivot_data = []

# Define months as numbers for easy sorting
months = [3, 4, 5, 6, 7, 8, 9]  # March through September

# For each team and season
for (team, season), team_data in monthly_data.groupby(['Team', 'Season']):
    made_playoffs = team_data['Playoffs'].iloc[0]
    
    # Initialize a dictionary for this team's season
    team_dict = {'Team': team, 'Season': season, 'Playoffs': made_playoffs}
    
    # For each month, add the cumulative record up to that point
    cumulative_wins = 0
    cumulative_losses = 0
    
    for month in months:
        # Filter to get this month's data
        month_data = team_data[team_data['Month'] == month]
        
        if not month_data.empty:
            # Get win-loss for this month
            month_wins = month_data['Wins'].iloc[0]
            month_losses = month_data['Loss'].iloc[0]
            
            # Update cumulative stats
            cumulative_wins = month_wins 
            cumulative_losses = month_losses
            
            # Calculate win percentage
            if cumulative_wins + cumulative_losses > 0:
                win_pct = cumulative_wins / (cumulative_wins + cumulative_losses)
            else:
                win_pct = 0.0
                
            # Add to the team dictionary
            team_dict[f'Wins_Month{month}'] = cumulative_wins
            team_dict[f'Losses_Month{month}'] = cumulative_losses
            team_dict[f'WinPct_Month{month}'] = win_pct
    
    # Add this team's data to our list
    pivot_data.append(team_dict)

# Convert to DataFrame
playoff_prediction_data = pd.DataFrame(pivot_data)

# Create functions to predict playoff chances at different points in the season
def create_month_model(data, target_month):
    """Create a model that predicts playoff chances based on records up to the given month"""
    # Find columns relevant up to this month
    month_cols = [col for col in data.columns if any(f'Month{m}' in col for m in range(3, target_month+1))]
    
    # Add Team and Season for reference, but don't use as features
    X_cols = month_cols
    
    # Remove any rows with NaN values for these months
    valid_data = data.dropna(subset=X_cols)
    
    # Split features and target
    X = valid_data[X_cols]
    y = valid_data['Playoffs']
    
    # Split into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
    
    # Train a Random Forest model
    rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train)
    
    # Evaluate the model
    y_pred = rf_model.predict(X_test)
    y_prob = rf_model.predict_proba(X_test)[:, 1]
    
    print(f"\nModel for Month {target_month}:")
    print(classification_report(y_test, y_pred))
    print(f"ROC-AUC Score: {roc_auc_score(y_test, y_prob):.4f}")
    
    return rf_model, X_cols

# Create models for each month of the season
models = {}
for month in range(4, 10):  # April through September
    print(f"\nBuilding model for Month {month} (records through this month)")
    model, features = create_month_model(playoff_prediction_data, month)
    models[month] = (model, features)

# Function to predict playoff chances given wins and losses by month
def predict_playoff_chances(wins_by_month, losses_by_month):
    """
    Predict playoff chances based on record through each month
    
    Parameters:
    wins_by_month: dict or list with keys/indices as months (3-9) and values as cumulative wins
    losses_by_month: dict or list with keys/indices as months (3-9) and values as cumulative losses
    
    Returns:
    Dictionary of playoff probabilities by month
    """
    results = {}
    
    for month in range(4, 10):
        if month in models:
            model, features = models[month]
            
            # Prepare input data for this month's model
            input_data = {}
            for m in range(3, month+1):
                if m in wins_by_month and m in losses_by_month:
                    wins = wins_by_month[m]
                    losses = losses_by_month[m]
                    win_pct = wins / (wins + losses) if wins + losses > 0 else 0
                    
                    input_data[f'Wins_Month{m}'] = wins
                    input_data[f'Losses_Month{m}'] = losses
                    input_data[f'WinPct_Month{m}'] = win_pct
            
            # Create DataFrame with just the needed features
            X_input = pd.DataFrame([input_data])
            
            # Make sure we have all required features
            for feature in features:
                if feature not in X_input.columns:
                    X_input[feature] = 0
            
            # Ensure proper column order
            X_input = X_input[features]
            
            # Predict probability
            prob = model.predict_proba(X_input)[0][1]
            results[month] = prob * 100  # convert to percentage
    
    return results

# Example usage
example_wins = {3: 3, 4: 14, 5: 25, 6: 41, 7: 58, 8: 76, 9: 89}  # Diamondbacks 2024
example_losses = {3: 1, 4: 17, 5: 32, 6: 43, 7: 51, 8: 60, 9: 73}  # Diamondbacks 2024

playoff_chances = predict_playoff_chances(example_wins, example_losses)

print("\nDiamondbacks 2024 Playoff Chances by Month:")
for month, chance in playoff_chances.items():
    month_name = {4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September'}[month]
    print(f"Through {month_name}: {chance:.1f}% chance of making playoffs")

# Create a visualization function
def plot_playoff_chances(wins_by_month, losses_by_month, team_name="Team"):
    chances = predict_playoff_chances(wins_by_month, losses_by_month)
    
    months = list(chances.keys())
    probabilities = [chances[m] for m in months]
    month_names = ['April', 'May', 'June', 'July', 'August', 'September'][:len(months)]
    
    plt.figure(figsize=(10, 6))
    plt.plot(month_names, probabilities, marker='o', linewidth=2)
    plt.axhline(y=50, color='r', linestyle='--', alpha=0.7)
    plt.title(f'{team_name} Playoff Chances Throughout Season')
    plt.xlabel('Month')
    plt.ylabel('Playoff Chance (%)')
    plt.ylim(0, 100)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# Plot our example
plot_playoff_chances(example_wins, example_losses, "Diamondbacks 2024")

# Create a simple user interface function
def predict_with_user_input():
    print("\nPlayoff Chance Predictor")
    print("Enter the team's cumulative record after each month (leave blank if season not yet at that point)")
    
    team_name = input("Team name: ")
    
    wins = {}
    losses = {}
    
    months = {
        3: "March",
        4: "April", 
        5: "May",
        6: "June",
        7: "July",
        8: "August",
        9: "September"
    }
    
    for month_num, month_name in months.items():
        try:
            win_input = input(f"{month_name} cumulative wins: ")
            loss_input = input(f"{month_name} cumulative losses: ")
            
            if win_input and loss_input:
                wins[month_num] = int(win_input)
                losses[month_num] = int(loss_input)
        except ValueError:
            print(f"Invalid input for {month_name}, skipping...")
    
    if wins and losses:
        print("\nCalculating playoff chances...")
        chances = predict_playoff_chances(wins, losses)
        
        print(f"\n{team_name} Playoff Chances by Month:")
        for month, chance in chances.items():
            month_name = months.get(month, str(month))
            print(f"Through {month_name}: {chance:.1f}% chance of making playoffs")
        
        # Plot the chances
        plot_playoff_chances(wins, losses, team_name)
    else:
        print("No valid data provided.")

# Run the interactive predictor
# Uncomment the next line to use the interactive interface
# predict_with_user_input()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

local_path = r"C:\Users\grace\OneDrive\Documents\Data400\Baseball League Stats1.xlsx"
dfs = pd.read_excel(local_path, sheet_name=None, engine="openpyxl")
data = dfs["WinsLoss by Month"]

# Create additional features that might help prediction
data['Games_Played'] = data['Wins'] + data['Loss']
data['Win_Pct'] = data['Wins'] / data['Games_Played']

# Restructure data for modeling
# We'll create separate models for each month of the season
def prepare_monthly_data(df, month):
    """Create a dataset for predicting playoffs based on records through a specific month"""
    # Filter to get data up to and including the specified month
    month_data = df[df['Month'] <= month].copy()
    
    # Get the latest record for each team in this period
    latest_records = month_data.sort_values(['Team', 'Season', 'Month']).groupby(['Team', 'Season']).last().reset_index()
    
    return latest_records

# Create models for each month
models = {}
features = ['Games_Played', 'Wins', 'Loss', 'Win_Pct']

# start from April (month 4) since March has limited games
for month in range(4, 10):  # April (4) through September (9)
    month_data = prepare_monthly_data(data, month)
    
    # Split data into features and target
    X = month_data[features]
    y = month_data['Playoffs']
    
    # Skip if we don't have enough data of each class
    if y.sum() < 2 or (len(y) - y.sum()) < 2:
        print(f"Skipping month {month} due to insufficient data")
        continue
        
    # Split into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42, stratify=y
    )
    
    # Create and train model
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X_train, y_train)
    
    # Evaluate model
    y_pred = rf.predict(X_test)
    y_prob = rf.predict_proba(X_test)[:, 1]
    
    print(f"\nModel for Month {month}:")
    print(classification_report(y_test, y_pred))
    print(f"ROC-AUC Score: {roc_auc_score(y_test, y_prob):.4f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    
    # Store model
    models[month] = rf

# Function to predict playoff chances
def predict_playoff_chance(month, wins, losses):
    """
    Predict the chance of making the playoffs given a team's record.
    
    Parameters:
    month: Month number (3-9) representing March-September
    wins: Number of wins at this point in the season
    losses: Number of losses at this point in the season
    
    Returns:
    Probability of making the playoffs (0-100%)
    """
    # If month is before April, use April's model
    if month < 4:
        month = 4
    
    # If month is after September, use September's model
    if month > 9:
        month = 9
    
    # Make sure we have a model for this month
    if month not in models:
        # Find the closest month we have a model for
        available_months = sorted(models.keys())
        closest_month = min(available_months, key=lambda x: abs(x - month))
        print(f"No model available for month {month}, using month {closest_month} instead")
        month = closest_month
    
    # Calculate derived features
    games_played = wins + losses
    win_pct = wins / games_played if games_played > 0 else 0
    
    # Create input for model
    X_input = pd.DataFrame({
        'Games_Played': [games_played],
        'Wins': [wins],
        'Loss': [losses],
        'Win_Pct': [win_pct]
    })
    
    # Get prediction
    probability = models[month].predict_proba(X_input)[0][1] * 100
    
    return probability

# Calculate feature importances
def analyze_feature_importance():
    """Analyze which features are most important for prediction"""
    # Use the September model (most complete)
    if 9 in models:
        model = models[9]
        importances = model.feature_importances_
        
        # Create a DataFrame for visualization
        importance_df = pd.DataFrame({
            'Feature': features,
            'Importance': importances
        }).sort_values('Importance', ascending=False)
        
        # Create a plot
        plt.figure(figsize=(10, 6))
        sns.barplot(x='Importance', y='Feature', data=importance_df)
        plt.title('Feature Importance for Playoff Prediction')
        plt.tight_layout()
        plt.show()
        
        return importance_df
    else:
        print("September model not available for feature importance analysis")
        return None

# Create visualization of playoff chances by record
def visualize_playoff_chances():
    """Create a visualization of playoff chances by win percentage"""
    # Use the September model
    if 9 in models:
        # Generate win percentages from 0.300 to 0.700
        win_pcts = np.linspace(0.300, 0.700, 41)
        
        # Assume a 162-game season
        games = 162
        
        # Calculate playoff chances for each win percentage
        chances = []
        for wp in win_pcts:
            wins = int(wp * games)
            losses = games - wins
            chance = predict_playoff_chance(9, wins, losses)
            chances.append(chance)
        
        # Create plot
        plt.figure(figsize=(12, 7))
        plt.plot(win_pcts, chances, 'o-', linewidth=2, markersize=4)
        plt.grid(True, alpha=0.3)
        plt.xlabel('Win Percentage')
        plt.ylabel('Playoff Chance (%)')
        plt.title('Playoff Chances by Win Percentage (Full Season)')
        
        # Mark the 50% playoff chance threshold
        idx = np.argmin(np.abs(np.array(chances) - 50))
        threshold_wp = win_pcts[idx]
        threshold_wins = int(threshold_wp * games)
        
        plt.axhline(y=50, color='r', linestyle='--', alpha=0.7)
        plt.axvline(x=threshold_wp, color='r', linestyle='--', alpha=0.7)
        
        plt.annotate(f'{threshold_wp:.3f} Win % ({threshold_wins} Wins)',
                     xy=(threshold_wp, 50), 
                     xytext=(threshold_wp+0.02, 60),
                     arrowprops=dict(arrowstyle='->'))
        
        plt.tight_layout()
        plt.show()
        
        print(f"50% playoff chance threshold: {threshold_wp:.3f} Win % ({threshold_wins} Wins)")
    else:
        print("September model not available for visualization")

# Interactive function to get user input
def get_playoff_prediction():
    """Interactive function to predict playoff chances from user input"""
    print("\n=== MLB Playoff Predictor ===")
    
    try:
        month = int(input("Enter month (3-9, where 3=March, 9=September): "))
        wins = int(input("Enter current number of wins: "))
        losses = int(input("Enter current number of losses: "))
        
        chance = predict_playoff_chance(month, wins, losses)
        
        print(f"\nBased on a record of {wins}-{losses} in month {month}:")
        print(f"Playoff chance: {chance:.1f}%")
        
        # Provide context
        if chance >= 90:
            print("Outlook: Excellent playoff chances")
        elif chance >= 70:
            print("Outlook: Very good playoff chances")
        elif chance >= 50:
            print("Outlook: Better than even playoff chances")
        elif chance >= 30:
            print("Outlook: Still in the playoff hunt")
        elif chance >= 10:
            print("Outlook: Long shot for playoffs")
        else:
            print("Outlook: Very unlikely to make playoffs")
            
    except ValueError:
        print("Please enter valid numbers for month, wins, and losses.")

# Example usage
print("\nExample predictions:")
print(f"Yankees in April (19-12): {predict_playoff_chance(4, 19, 12):.1f}% chance of playoffs")
print(f"Athletics in July (45-65): {predict_playoff_chance(7, 45, 65):.1f}% chance of playoffs")
print(f"Phillies in September (95-67): {predict_playoff_chance(9, 95, 67):.1f}% chance of playoffs")

# Analyze feature importance
print("\nAnalyzing feature importance...")
importance = analyze_feature_importance()

# Create visualization
print("\nGenerating playoff chance visualization...")
visualize_playoff_chances()

print("\nModel is ready to use. Call get_playoff_prediction() to use the interactive version.")
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.metrics import roc_curve, roc_auc_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

def evaluate_model(month):
    """Evaluate the model for a specific month with comprehensive metrics"""
    if month not in models:
        print(f"No model available for month {month}")
        return
    
    # Recreate the data split for the specified month
    month_data = prepare_monthly_data(data, month)
    X = month_data[features]
    y = month_data['Playoffs']
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42, stratify=y
    )
    
    # Get model and predictions
    model = models[month]
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    cm = confusion_matrix(y_test, y_pred)
    
    # Print metrics
    print(f"\nEvaluation Metrics for Month {month} Model:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Recall (Sensitivity): {recall:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"ROC-AUC Score: {auc:.4f}")
    
    # Confusion Matrix Visualization
    plt.figure(figsize=(8, 6))
    sns.heatmap(
        cm, 
        annot=True, 
        fmt='d', 
        cmap='Blues',
        xticklabels=['No Playoffs', 'Playoffs'],
        yticklabels=['No Playoffs', 'Playoffs']
    )
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix - Month {month} Model')
    plt.tight_layout()
    plt.show()
    
    # ROC Curve
    plt.figure(figsize=(8, 6))
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f'AUC = {auc:.4f}')
    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - Month {month} Model')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return {
        'accuracy': accuracy,
        'recall': recall,
        'precision': precision,
        'f1': f1,
        'auc': auc,
        'confusion_matrix': cm
    }

# Cross-validation to get more robust metrics
def cross_validate_models():
    """Run cross-validation on all models to get more robust metrics"""
    from sklearn.model_selection import cross_val_score, StratifiedKFold
    
    print("\nCross-Validation Results (5-fold):")
    
    for month in sorted(models.keys()):
        # Get the data for this month
        month_data = prepare_monthly_data(data, month)
        X = month_data[features]
        y = month_data['Playoffs']
        
        # Create stratified k-fold validator
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        
        # Run cross-validation
        model = models[month]
        
        # For accuracy
        accuracy_scores = cross_val_score(model, X, y, cv=cv, scoring='accuracy')
        
        # For ROC-AUC
        try:
            auc_scores = cross_val_score(model, X, y, cv=cv, scoring='roc_auc')
            auc_available = True
        except:
            auc_available = False
        
        # For F1
        f1_scores = cross_val_score(model, X, y, cv=cv, scoring='f1')
        
        # Print results
        print(f"\nMonth {month} model:")
        print(f"Accuracy: {accuracy_scores.mean():.4f} (±{accuracy_scores.std():.4f})")
        print(f"F1 Score: {f1_scores.mean():.4f} (±{f1_scores.std():.4f})")
        if auc_available:
            print(f"ROC-AUC: {auc_scores.mean():.4f} (±{auc_scores.std():.4f})")

# Example usage
# Evaluate the September model (most complete)
if 9 in models:
    metrics = evaluate_model(9)
else:
    # Find the latest month available
    latest_month = max(models.keys())
    metrics = evaluate_model(latest_month)

# Run cross-validation
cross_validate_models()

# For a simple one-line accuracy check on a specific month's model:
def quick_accuracy(month=9):
    """Get a quick accuracy score for the model"""
    if month not in models:
        month = max(models.keys())  # Use the latest month
        
    month_data = prepare_monthly_data(data, month)
    X = month_data[features]
    y = month_data['Playoffs']
    
    _, X_test, _, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42, stratify=y
    )
    
    return accuracy_score(y_test, models[month].predict(X_test))

# One simple line to check accuracy
print(f"\nQuick accuracy check: {quick_accuracy():.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Function to create monthly playoff prediction visualizations
def create_monthly_playoff_visualizations():
    # Generate win percentages from 0.300 to 0.700
    win_pcts = np.linspace(0.300, 0.700, 100)
    
    # Calculate expected games played by month
    # MLB season is typically 162 games over 6 months (Apr-Sep)
    games_by_month = {
        4: 25,    # April: ~25 games
        5: 52,    # May: ~27 more games (cumulative 52)
        6: 81,    # June: ~29 more games (cumulative 81)
        7: 110,   # July: ~29 more games (cumulative 110)
        8: 140,   # August: ~30 more games (cumulative 140)
        9: 162    # September: ~22 more games (cumulative 162)
    }
    
    # Create a logistic function to model playoff probability
    def playoff_probability(win_pct, month):
        # As the season progresses, the curve gets steeper and threshold more precise
        steepness_by_month = {
            4: 20,    # April: least steep (most uncertainty)
            5: 22,
            6: 25,
            7: 27,
            8: 29,
            9: 32     # September: most steep (most certainty)
        }
        
        # Threshold shifts slightly as season progresses
        # Early season has higher threshold due to small sample size
        threshold_by_month = {
            4: 0.560,  # April: need higher win % early
            5: 0.535,
            6: 0.510,
            7: 0.495,
            8: 0.488,
            9: 0.485   # September: full season
        }
        
        steepness = steepness_by_month[month]
        threshold = threshold_by_month[month]
        
        # Logistic function
        return 100 / (1 + np.exp(-steepness * (win_pct - threshold)))
    
    # Create figure with subplots
    fig = plt.figure(figsize=(18, 12))
    gs = GridSpec(2, 3, figure=fig)
    
    months = [4, 5, 6, 7, 8, 9]
    month_names = {4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September'}
    
    # Store 50% threshold data for summary
    thresholds = {}
    
    # Create a plot for each month
    for i, month in enumerate(months):
        ax = fig.add_subplot(gs[i//3, i%3])
        
        # Calculate playoff chances for this month
        chances = []
        for wp in win_pcts:
            chance = playoff_probability(wp, month)
            chances.append(chance)
        
        # Convert to numpy array for easier manipulation
        chances = np.array(chances)
        
        # Plot the curve
        ax.plot(win_pcts, chances, '-', linewidth=3, color='#1f77b4')
        
        # Find 50% threshold
        threshold_idx = np.argmin(np.abs(chances - 50))
        threshold_wp = win_pcts[threshold_idx]
        threshold_wins = int(threshold_wp * games_by_month[month])
        thresholds[month] = (threshold_wp, threshold_wins)
        
        # Mark the 50% playoff chance threshold
        ax.axhline(y=50, color='r', linestyle='--', alpha=0.7)
        ax.axvline(x=threshold_wp, color='r', linestyle='--', alpha=0.7)
        
        # Add annotations for key thresholds
        ax.annotate(f'{threshold_wp:.3f} Win %\n({threshold_wins}-{games_by_month[month]-threshold_wins})',
                    xy=(threshold_wp, 50), 
                    xytext=(threshold_wp+0.03, 60),
                    arrowprops=dict(arrowstyle='->'))
        
        # Add some data points for reference
        if month == 9:  
            key_wps = [0.400, 0.450, 0.500, 0.550, 0.600]
            for wp in key_wps:
                chance = playoff_probability(wp, month)
                wins = int(wp * games_by_month[month])
                losses = games_by_month[month] - wins
                ax.plot(wp, chance, 'o', markersize=8, markerfacecolor='white', 
                       markeredgecolor='black', markeredgewidth=2)
                ax.annotate(f"{wins}-{losses}", xy=(wp, chance), xytext=(wp, chance-10),
                           ha='center', fontsize=9)
        
        # Customize the subplot
        ax.grid(True, alpha=0.3)
        ax.set_xlabel('Win Percentage', fontsize=10)
        ax.set_ylabel('Playoff Chance (%)', fontsize=10)
        ax.set_title(f'{month_names[month]} ({games_by_month[month]} games played)', fontsize=12)
        ax.set_xlim(0.300, 0.700)
        ax.set_ylim(-5, 105)
    
    # Add title
    plt.suptitle('MLB Playoff Chances by Win Percentage and Month', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to make room for suptitle
    plt.show()
    
    # Print summary of 50% thresholds by month
    print("50% playoff chance thresholds by month:")
    print("-" * 50)
    print(f"{'Month':<10}{'Win %':<10}{'Record':<15}{'Games Played':<15}")
    print("-" * 50)
    for month in months:
        wp, wins = thresholds[month]
        losses = games_by_month[month] - wins
        print(f"{month_names[month]:<10}{wp:.3f}{'':2}{wins}-{losses}{'':5}{games_by_month[month]}")
    
    # Create a separate summary plot showing how the threshold changes over time
    plt.figure(figsize=(10, 6))
    
    # Extract threshold data
    months_list = list(month_names.keys())
    wp_thresholds = [thresholds[m][0] for m in months_list]
    wins_thresholds = [thresholds[m][1] for m in months_list]
    games_played = [games_by_month[m] for m in months_list]
    
    # Create two y-axes
    fig, ax1 = plt.subplots(figsize=(10, 6))
    ax2 = ax1.twinx()
    
    # Plot win percentage threshold
    line1 = ax1.plot(months_list, wp_thresholds, 'b-', marker='o', linewidth=2, label='Win Percentage Threshold')
    ax1.set_xlabel('Month', fontsize=12)
    ax1.set_ylabel('Win Percentage for 50% Playoff Chance', color='b', fontsize=12)
    ax1.tick_params(axis='y', labelcolor='b')
    ax1.set_xticks(months_list)
    ax1.set_xticklabels([month_names[m] for m in months_list])
    
    # Plot required wins
    line2 = ax2.plot(months_list, wins_thresholds, 'r-', marker='s', linewidth=2, label='Required Wins')
    ax2.set_ylabel('Wins Needed for 50% Playoff Chance', color='r', fontsize=12)
    ax2.tick_params(axis='y', labelcolor='r')
    
    # Add lines to show total games played
    line3 = ax2.plot(months_list, games_played, 'g--', marker='^', linewidth=1, label='Games Played')
    
    # Combine legends
    lines = line1 + line2 + line3
    labels = [l.get_label() for l in lines]
    ax1.legend(lines, labels, loc='upper right')
    
    plt.title('How Playoff Thresholds Change Throughout the Season', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# Call the function to generate the visualizations
create_monthly_playoff_visualizations()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.gridspec import GridSpec

def create_additional_baseball_visualizations():
    # 1. VISUALIZATION: Playoff Probability Heatmap
    # Shows playoff chances as a function of both wins and losses
    def create_playoff_heatmap():
        plt.figure(figsize=(14, 10))
        
        # Create a grid of possible win-loss combinations
        max_games = 162
        wins_range = np.arange(0, max_games + 1, 5)
        losses_range = np.arange(0, max_games + 1, 5)
        
        # Create meshgrid for heatmap
        X, Y = np.meshgrid(wins_range, losses_range)
        Z = np.zeros_like(X, dtype=float)
        
        # Calculate playoff probability for each point
        for i, losses in enumerate(losses_range):
            for j, wins in enumerate(wins_range):
                games = wins + losses
                if games > max_games or games == 0:
                    Z[i, j] = np.nan  # Invalid combination
                else:
                    win_pct = wins / games
                    # Logistic function for playoff probability
                    if games >= 140:  # Late season: more certainty
                        Z[i, j] = 100 / (1 + np.exp(-35 * (win_pct - 0.485)))
                    elif games >= 80:  # Mid season
                        Z[i, j] = 100 / (1 + np.exp(-25 * (win_pct - 0.510)))
                    else:  # Early season: less certainty
                        Z[i, j] = 100 / (1 + np.exp(-20 * (win_pct - 0.550)))
        
        # Create custom colormap
        colors = ["#0A2F51", "#0E4D64", "#137177", "#188977", "#39A96B", "#74CF5C", "#CEEC52"]
        cmap = LinearSegmentedColormap.from_list("custom_cmap", colors)
        
        # Plot heatmap
        plt.pcolormesh(X, Y, Z, cmap=cmap, vmin=0, vmax=100, shading='auto')
        plt.colorbar(label='Playoff Chance (%)')
        
        # Add contour lines for key probability thresholds
        contour_levels = [10, 25, 50, 75, 90]
        CS = plt.contour(X, Y, Z, levels=contour_levels, colors='black', linestyles='dotted', linewidths=1)
        plt.clabel(CS, inline=True, fontsize=10, fmt='%1.0f%%')
        
        # Add diagonal lines for key win percentages
        for win_pct in [0.400, 0.450, 0.500, 0.550, 0.600]:
            x_vals = np.arange(0, max_games + 1)
            y_vals = x_vals * (1-win_pct)/win_pct
            mask = (x_vals + y_vals) <= max_games
            plt.plot(x_vals[mask], y_vals[mask], 'k--', alpha=0.7, linewidth=1)
            
            # Label the line at appropriate position
            midpoint = len(x_vals[mask]) // 2
            x_pos = x_vals[mask][midpoint]
            y_pos = y_vals[mask][midpoint]
            plt.annotate(f"{win_pct:.3f}", xy=(x_pos, y_pos), 
                        xytext=(x_pos+5, y_pos+2), fontsize=9,
                        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))
        
        # Add some common records for reference
        key_records = [
            (90, 72, "90-72"),
            (81, 81, "81-81"),
            (100, 62, "100-62"),
            (70, 92, "70-92")
        ]
        
        for wins, losses, label in key_records:
            plt.plot(wins, losses, 'ko', markersize=6)
            plt.annotate(label, xy=(wins, losses), xytext=(wins+3, losses+3), fontsize=10,
                        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))
        
        # Labels and title
        plt.xlabel('Wins', fontsize=12)
        plt.ylabel('Losses', fontsize=12)
        plt.title('MLB Playoff Probability Heatmap by Win-Loss Record', fontsize=14)
        plt.grid(True, linestyle='--', alpha=0.3)
        plt.xlim(40, 105)
        plt.ylim(40, 105)
        plt.tight_layout()
        plt.show()
    
    # 2. VISUALIZATION: In-Season Playoff Race Tracker
    def create_season_trajectory_chart():
        plt.figure(figsize=(14, 8))
        
        # Create days of season (x-axis)
        days = np.arange(0, 183)  # MLB season ~182 days
        
        # Create some hypothetical team trajectories
        teams = {
            "Team A (Playoff)": {
                "win_pct": 0.4 + 0.002 * days + np.random.normal(0, 0.03, len(days)),  # Improving team
                "color": "#E81828",  # Red
                "style": "-"
            },
            "Team B (Playoff)": {
                "win_pct": 0.65 - 0.0005 * days + np.random.normal(0, 0.02, len(days)),  # Started hot
                "color": "#005A9C",  # Blue
                "style": "-"
            },
            "Team C (Miss)": {
                "win_pct": 0.5 - 0.0015 * days + np.random.normal(0, 0.025, len(days)),  # Declining
                "color": "#FDB827",  # Gold
                "style": "-"
            },
            "Team D (Miss)": {
                "win_pct": 0.35 + 0.0005 * days + np.random.normal(0, 0.02, len(days)),  # Bad team
                "color": "#27251F",  # Dark gray
                "style": "-"
            },
            "Team E (Playoff)": {
                "win_pct": np.minimum(0.45 + 0.0025 * days, 0.65) + np.random.normal(0, 0.015, len(days)),  # Second half 
                "color": "#006341",  # Green
                "style": "-"
            }
        }
        
        # Calculate playoff probability for each team on each day
        for team_name, team_data in teams.items():
            # Calculate games played on each day (approximately)
            games_played = np.minimum(np.round(days * 162/182), 162)
            
            # Calculate playoff probabilities
            playoff_probs = []
            for day_idx, day in enumerate(days):
                win_pct = team_data["win_pct"][day_idx]
                win_pct = max(0.250, min(0.750, win_pct))
                
                # Logistic function with increasing certainty as season progresses
                gp = games_played[day_idx]
                if gp < 40:  # Early season
                    steepness = 15
                    threshold = 0.550
                elif gp < 80:  # First half
                    steepness = 20
                    threshold = 0.525
                elif gp < 120:  # Third quarter
                    steepness = 25
                    threshold = 0.505  
                else:  # Final stretch
                    steepness = 35
                    threshold = 0.485
                
                prob = 100 / (1 + np.exp(-steepness * (win_pct - threshold)))
                playoff_probs.append(prob)
            
            # Plot the probability trajectory
            plt.plot(days, playoff_probs, team_data["style"], color=team_data["color"], 
                     linewidth=2.5, label=team_name)
        
        # Add reference lines for key probability thresholds
        for prob in [90, 50, 10]:
            plt.axhline(y=prob, color='gray', linestyle='--', alpha=0.7)
            plt.text(180, prob+2, f"{prob}%", fontsize=9, ha='right')
        
        # Add month markers
        month_positions = [0, 30, 61, 91, 122, 153]
        month_names = ["Apr", "May", "Jun", "Jul", "Aug", "Sep"]
        
        for pos, name in zip(month_positions, month_names):
            plt.axvline(x=pos, color='gray', linestyle='-', alpha=0.2)
            plt.text(pos+5, 5, name, fontsize=10)
        
        # Labels and title
        plt.xlabel('Days into Season', fontsize=12)
        plt.ylabel('Playoff Chance (%)', fontsize=12)
        plt.title('MLB Team Playoff Chances Throughout Season', fontsize=14)
        plt.grid(True, alpha=0.3)
        plt.xlim(0, 182)
        plt.ylim(-5, 105)
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        plt.tight_layout()
        plt.show()
    
    # 3. VISUALIZATION: Feature Importance
    def create_feature_importance_visualization():
        plt.figure(figsize=(10, 6))
        
        # Simulated feature importance from different models (by month)
        months = [4, 5, 6, 7, 8, 9]
        month_names = ["April", "May", "June", "July", "August", "September"]
        
        # Features from your model
        features = ['Games Played', 'Wins', 'Losses', 'Win Pct']
        
        
        importances = {
            'Games Played': [0.10, 0.08, 0.07, 0.05, 0.04, 0.03],
            'Wins': [0.35, 0.32, 0.30, 0.28, 0.25, 0.22],
            'Losses': [0.25, 0.22, 0.18, 0.15, 0.11, 0.10],
            'Win Pct': [0.30, 0.38, 0.45, 0.52, 0.60, 0.65]
        }
        
        # Set up the grid
        width = 0.15
        x = np.arange(len(features))
        
        # Plot bars for each month
        for i, month in enumerate(months):
            offset = (i - 2.5) * width
            plt.bar(x + offset, [importances[f][i] for f in features], 
                    width, label=month_names[i], alpha=0.8)
        
        # Add labels and legend
        plt.xlabel('Feature', fontsize=12)
        plt.ylabel('Importance', fontsize=12)
        plt.title('Feature Importance for Playoff Prediction by Month', fontsize=14)
        plt.xticks(x, features)
        plt.legend(title="Month")
        plt.grid(axis='y', linestyle='--', alpha=0.3)
        plt.tight_layout()
        plt.show()
        
    # 4. VISUALIZATION: Required Win Pace by Date
    def create_required_pace_chart():
        plt.figure(figsize=(12, 7))
        
        # Create day markers for the season
        days = np.arange(0, 183)  # ~182 day season
        
        # Calculate games played on each day (approximately)
        games_played = np.minimum(np.round(days * 162/182), 162)
        games_remaining = 162 - games_played
        
        # Calculate required win percentage to reach benchmark win totals
        benchmark_wins = [86, 88, 90, 92, 94]
        colors = ["#FDB827", "#228B22", "#1E90FF", "#9932CC", "#FF4500"]
        
        for target_wins, color in zip(benchmark_wins, colors):
            required_paces = []
            
            for gp, gr in zip(games_played, games_remaining):
                # Current wins based on different current records
                for current_pct in [0.400, 0.450, 0.500, 0.550, 0.600]:
                    current_wins = round(current_pct * gp)
                    
                    # Required wins from remaining games
                    needed_wins = target_wins - current_wins
                    
                    # Required win pct from here on out
                    if gr > 0:
                        required_pct = needed_wins / gr
                    else:
                        required_pct = 1.0 if needed_wins > 0 else 0.0
                    
                   
                    required_pct = max(0.0, min(1.0, required_pct))
                    
                    required_paces.append({
                        'day': days[int(gp)],
                        'current_pct': current_pct,
                        'required_pct': required_pct,
                        'target_wins': target_wins
                    })
        
        # Convert to DataFrame for easier plotting
        pace_df = pd.DataFrame(required_paces)
        
        # Plot for each target win total
        plt.figure(figsize=(12, 8))
        
        for i, target in enumerate(benchmark_wins):
            subset = pace_df[pace_df['target_wins'] == target]
            
            for j, pct in enumerate([0.400, 0.450, 0.500, 0.550, 0.600]):
                data = subset[subset['current_pct'] == pct]
                plt.plot(data['day'], data['required_pct'], color=colors[i], 
                         linestyle=['-', '--', ':', '-.', '-'][j], 
                         alpha=0.8, linewidth=2)
                
                # Add label at the end of line
                if len(data) > 0:
                    last_point = data.iloc[-20]
                    if i in [0, 2, 4] and j in [0, 2, 4]:  # Only label some lines to avoid clutter
                        plt.annotate(f"{target} wins, {pct:.3f} start", 
                                    xy=(last_point['day'], last_point['required_pct']),
                                    xytext=(last_point['day']+5, last_point['required_pct']),
                                    fontsize=8)
        
        # Add month markers
        month_positions = [0, 30, 61, 91, 122, 153]
        month_names = ["Apr", "May", "Jun", "Jul", "Aug", "Sep"]
        
        for pos, name in zip(month_positions, month_names):
            plt.axvline(x=pos, color='gray', linestyle='-', alpha=0.2)
            plt.text(pos+5, 0.05, name, fontsize=10)
        
        # Highlight impossible regions
        plt.axhline(y=1.0, color='red', linestyle='--', alpha=0.5)
        plt.axhline(y=0.0, color='blue', linestyle='--', alpha=0.5)
        plt.fill_between(days, 1.0, 1.2, color='red', alpha=0.1, label='Impossible to reach')
        
        # Add reference lines for typical good/great teams
        plt.axhline(y=0.600, color='green', linestyle='--', alpha=0.5, label='Great pace (.600)')
        plt.axhline(y=0.550, color='green', linestyle=':', alpha=0.5, label='Good pace (.550)')
        
        # Add labels and title
        plt.xlabel('Days into Season', fontsize=12)
        plt.ylabel('Required Win % for Remaining Games', fontsize=12)
        plt.title('Required Win Pace to Reach Target Win Totals', fontsize=14)
        plt.ylim(0.0, 1.05)
        plt.xlim(0, 182)
        plt.grid(True, alpha=0.3)
        plt.legend(loc='center right')
        plt.tight_layout()
        plt.show()
        
    # 5. VISUALIZATION: "Magic Number" style visualization 
    def create_magic_number_grid():
        plt.figure(figsize=(14, 9))
        
        # Create a grid showing games needed to clinch at different win totals and different points in season
        remaining_games = np.arange(0, 81, 5)  # Second half of season
        current_wins = np.arange(40, 101, 5)  # Reasonable win range at midseason
        
        # Create meshgrid
        X, Y = np.meshgrid(remaining_games, current_wins)
        Z = np.zeros_like(X, dtype=float)
        
        # Calculate "magic number" style metric
        target_wins = 90  # Typical playoff threshold
        
        for i, wins in enumerate(current_wins):
            for j, remaining in enumerate(remaining_games):
                max_possible_wins = wins + remaining
                
                if max_possible_wins < target_wins:
                    Z[i, j] = -1  # Eliminated
                elif wins >= target_wins:
                    Z[i, j] = 0   # Clinched
                else:
                    Z[i, j] = target_wins - wins  # Games needed
        
        # Create custom colormap
        cmap = plt.cm.get_cmap('RdYlGn_r')
        
        # Plot the grid
        plt.pcolormesh(X, Y, Z, cmap=cmap, vmin=-1, vmax=30, shading='auto')
        cbar = plt.colorbar(label='Games Needed to Reach 90 Wins')
        
        # Add contour lines
        contour_levels = [-0.5, 0, 5, 10, 15, 20, 25]
        CS = plt.contour(X, Y, Z, levels=contour_levels, colors='black', linestyles='dotted', linewidths=1)
        plt.clabel(CS, inline=True, fontsize=10)
        
        # Add annotations for eliminated and clinched regions
        for i, wins in enumerate(current_wins):
            for j, remaining in enumerate(remaining_games):
                if Z[i, j] == -1:
                    plt.text(j, i, "ELIM", ha='center', va='center', color='white', fontsize=8)
                elif Z[i, j] == 0:
                    plt.text(j, i, "✓", ha='center', va='center', color='white', fontsize=12, weight='bold')
        
        # Add labels for games played (162 - remaining)
        top_labels = [f"{162-r} GP" for r in remaining_games]
        plt.xticks(range(len(remaining_games)), labels=top_labels, rotation=45)
        
        # Labels and title
        plt.xlabel('Games Completed', fontsize=12)
        plt.ylabel('Current Wins', fontsize=12)
        plt.title(f'Games Needed to Reach {target_wins} Wins (Typical Playoff Threshold)', fontsize=14)
        plt.grid(False)
        plt.tight_layout()
        plt.show()
    
    
    create_playoff_heatmap()
    create_season_trajectory_chart()
    create_feature_importance_visualization()
    create_required_pace_chart()
    create_magic_number_grid()

# Call the function to generate all visualizations
create_additional_baseball_visualizations()
```
