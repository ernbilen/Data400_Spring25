### Exploratory Data Analysis

```python
from tabulate import tabulate
import pandas as pd
```

```python
# Read a specific sheet into a DataFrame
dataTSA = pd.read_excel('TSAData_2024-25.xlsx', sheet_name='TSA')
dataTSA.head()
```

```python
# Get value counts
# Convert DataFrame to Markdown
print(tabulate(pd.DataFrame(dataTSA[['State', 'Subject Matter']].value_counts()),\
     headers='keys', tablefmt='pipe'))
```

```python
# Get summary by grouping
dataTSA.groupby(['Subject Matter', 'State']).count().head(15)
```
